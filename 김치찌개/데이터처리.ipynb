```python
import pandas as pd

gang = pd.read_csv('서울_강남구_음식점_댓글수집.csv')
gan = pd.read_csv('서울_관악구_음식점_댓글수집.csv')
ap = pd.read_csv('서울_압구정_음식점_댓글수집.csv')
yee = pd.read_csv('서울_이태원_음식점_댓글수집.csv')
chung = pd.read_csv('서울_청담동_음식점_댓글수집.csv')
hong = pd.read_csv('서울_홍대_음식점_댓글수집.csv')
```

```python
gang
```

```python
gang=gang.drop_duplicates(keep='first')
gan=gan.drop_duplicates(keep='first')
ap=ap.drop_duplicates(keep='first')
yee=yee.drop_duplicates(keep='first')
chung=chung.drop_duplicates(keep='first')
hong=hong.drop_duplicates(keep='first')
```

```python
soul = pd.concat([gang, gan, ap, yee, chung, hong], axis=0)
soul
```

```python
# 숫자 인덱스 사용
soul.reset_index(inplace=True)
soul
```

```python
soul = soul.drop('index', axis=1)
soul
```

```python
soul['작성 날짜'][0].split('.')[0]
```

```python
year, month, day = [], [], []

for i in soul['작성 날짜']:
    a=i.split('.')
    if 3 == len(a):
        year.append('2024')
        month.append(a[0])
        day.append(a[1])
    elif 5 == len(a):
        year.append('20'+a[0])
        month.append(a[1])
        day.append(a[2])

len(year), len(month), len(day)
```

```python
li = []

for i in range(len(year)):
   li.append (f'{year[i]}-{int(month[i]):02}-{int(day[i]):02}')
li
```

```python
soul['날짜'] = li
soul.head()
```

```python
soul = soul.drop('작성 날짜', axis=1)
soul
```

```python
weather = pd.read_csv('기상청_강수량_2019_2024.csv')
temp = pd.read_csv('기상청_온도_일교차_2019_2024.csv')
sunshine = pd.read_csv('기상청_일조일사_2019_2024.csv')

columns = ['지점번호', '지점명', '일시', '강수량', 'ㅁ', 'ㅍ', 'ㅁ1']
weather.columns = columns
columns = ['지점',	'지점명', '일시', '일조합(hr)', '일조율(%)', '일사합(MJ/m2)']
sunshine.columns = columns
columns = ['지점',	'지점명', '일시', '평균기온(℃)', '최고기온(℃)', '최고기온시각',	'최저기온(℃)', '최저기온시각', '일교차']
temp.columns = columns
```

```python
sunshine
```

```python
weather = weather.fillna(0)
weather
```

```python
temp
```

```python
def w_col(string):
    li=[]
    for i in soul['날짜']:
        a=weather.loc[weather['일시']==i]
        li.append(a[string].values[0])
    return li
```

```python
def t_col(string):
    li=[]
    for i in soul['날짜']:
        a=temp.loc[temp['일시']==i]
        li.append(a[string].values[0])
    return li
```

```python
def s_col(string):
    li=[]
    for i in soul['날짜']:
        a=sunshine.loc[sunshine['일시']==i]
        li.append(a[string].values[0])
    return li
```

```python
soul['강수량'] = w_col('강수량')
soul['최고기온(℃)'] = t_col('최고기온(℃)')
soul['평균기온(℃)'] = t_col('평균기온(℃)')
soul['최저기온(℃)'] = t_col('최저기온(℃)')
soul['일조합(hr)'] = s_col('일조합(hr)')
soul['일조율(%)'] = s_col('일조율(%)')
soul['일사합(MJ/m2)'] = s_col('일사합(MJ/m2)')
soul
```

```python
li = []

for i in soul['날짜']:
    month = int(i.split('-')[1])
    if month in [3, 4, 5]:
        li.append('봄')
    elif month in [6, 7, 8]:
        li.append('여름')
    elif month in [9, 10, 11]:
        li.append('가을')
    else:
        li.append('겨울')

li
```

```python
soul['계절'] = li
soul
```
